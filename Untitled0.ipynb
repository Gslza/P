```python
#nama : Gusli Yanza
#nim : 8030230056
#kelas : 02PK4

import matplotlib.pyplot as plt
import numpy as np

# Buat koordinat titik-titik pentagon dengan rotasi agar rata di bawah
angles = np.linspace(0, 2 * np.pi, 6)[:-1] + np.pi / 2  # Tambah rotasi 90 derajat
radius = 1
x = radius * np.cos(angles)
y = radius * np.sin(angles)

# Warna untuk tiap sisi (searah jarum jam, mulai dari sisi atas kiri)
colors = ['blue', 'magenta', 'red', 'orange', 'lime']

# Gambar pentagon sisi per sisi
fig, ax = plt.subplots()
for i in range(5):
    x_vals = [x[i], x[(i + 1) % 5]]
    y_vals = [y[i], y[(i + 1) % 5]]
    ax.plot(x_vals, y_vals, color=colors[i], linewidth=5)

# Pengaturan tampilan
ax.set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
# Nama : Gusli Yanza
# NIM  : 8030230056
# Kelas: 02PK4

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

# Fungsi bantu menggambar bentuk berdasarkan titik
def draw_shape(points, color):
    shape = np.array(points)
    shape = np.vstack([shape, shape[0]])  # tutup bentuk
    ax.plot(shape[:, 0], shape[:, 1], color=color, linewidth=3)

# 1. Segitiga
triangle = [[0, 2], [-1, 0], [1, 0]]
draw_shape(triangle, 'red')

# 2. Belah ketupat
diamond = [[3, 0], [4, 1], [5, 0], [4, -1]]
draw_shape(diamond, 'purple')

# 3. Trapesium
trapezoid = [[6, 0], [7.5, 0], [7, -2], [6.5, -2]]
draw_shape(trapezoid, 'green')

# 4. Panah ke bawah
arrow = [[3.5, -3], [5.5, -3], [5.5, -4], [6, -4], [4.5, -6], [3, -4], [3.5, -4]]
draw_shape(arrow, 'orange')

# Pengaturan tampilan
ax.set_aspect('equal')
plt.axis('off')
plt.title("Bentuk Geometri: Segitiga, Belah Ketupat, Trapesium, Panah")
plt.show()
```

```python
#nama : Gusli Yanza
#nim : 8030230056
#kelas : 02PK4

import matplotlib.pyplot as plt
import numpy as np

# Jumlah sisi (oktagon)
n = 8
radius = 1

# Hitung koordinat titik-titik sudut oktagon
angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
x = radius * np.cos(angles)
y = radius * np.sin(angles)

# Gambar
fig, ax = plt.subplots(figsize=(6, 6))

# Hubungkan setiap titik dengan semua titik lainnya (pola bintang)
for i in range(n):
    for j in range(i+1, n):
        ax.plot([x[i], x[j]], [y[i], y[j]], color='slateblue', linewidth=1)

# Pengaturan visual
ax.set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
#nama : Gusli Yanza
#nim : 8030230056
#kelas : 02PK4

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

def star_points(center, radius1, radius2, num_points=5):
    """Generate the x, y coordinates of a star."""
    angles = np.linspace(0, 2 * np.pi, num_points * 2, endpoint=False)
    points = []
    for i, angle in enumerate(angles):
        r = radius1 if i % 2 == 0 else radius2
        x = center[0] + r * np.cos(angle + np.pi/2)
        y = center[1] + r * np.sin(angle + np.pi/2)
        points.append([x, y])
    return points

def draw_converse_logo():
    fig, ax = plt.subplots(figsize=(6,6))
    ax.set_xlim(-7, 7)
    ax.set_ylim(-6, 6)
    ax.set_aspect('equal')
    ax.axis('off')

    # Chevron (>)
    chevron = Polygon([
        [0.5, 5],
        [4.5, 1],
        [0.5, -3],
        [2.3, -3],
        [6.3, 1],
        [2.3, 5]
    ], closed=True, color='black')
    ax.add_patch(chevron)

    # Star (5-pointed)
    star = Polygon(star_points(center=(-3, 0.7), radius1=2.2, radius2=0.9, num_points=5),
                   closed=True, color='black')
    ax.add_patch(star)

    plt.show()

if __name__ == "__main__":
    draw_converse_logo()
```

```python
#nama : Gusli Yanza
#nim : 8030230056
#kelas : 02PK4

import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

def draw_adobe_logo():
    fig, ax = plt.subplots(figsize=(7,7))
    ax.set_xlim(-2, 8)
    ax.set_ylim(-2, 8)
    ax.set_aspect('equal')
    ax.axis('off')

    # Left wing (gunakan koordinat sesuai grid gambar)
    left_poly = Polygon([
        [0, 0],    # bawah kiri
        [0, 6],    # atas kiri
        [2, 6],    # atas kanan
        [0, 0]     # bawah kanan
    ], closed=True, color='red', alpha=0.6)
    ax.add_patch(left_poly)

    # Right wing
    right_poly = Polygon([
        [6, 0],    # bawah kiri
        [4, 6],    # atas kiri
        [6, 6],    # atas kanan
        [6, 0]     # bawah kanan
    ], closed=True, color='red', alpha=0.6)
    ax.add_patch(right_poly)

    # Middle "A"
    middle_poly = Polygon([
        [1, 1],    # kiri bawah
        [2.6, 3.5],# kiri tengah
        [3, 4],  # puncak atas
        [4.2, 1],  # kanan bawah
        [3.4, 1],  # sudut bawah menjorok kiri
        [3, 2.4],  # tengah bawah dalam
        [2.6, 1]    # kiri bawah
    ], closed=True, color='red', alpha=0.6)
    ax.add_patch(middle_poly)

    plt.text(3.5, 7, "Logo Adobe", fontsize=14, ha='center', color='gray')
    plt.show()

if __name__ == "__main__":
    draw_adobe_logo()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_microsoft_office_logo():
    fig, ax = plt.subplots(figsize=(4,5))
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 15)
    ax.axis('off')

    # Main red polygon (the "O" shape)
    points = [
        (2, 2),     # Bottom-left
        (4.5, 13),  # Top-left
        (10, 12),   # Top-right
        (10, 3.5),  # Bottom-right
        (5.5, 2.5), # Inner-bottom-right
        (5.5, 10),  # Inner-top-right
        (4.5, 10.2),# Inner-top-left
        (4.5, 3.7), # Inner-bottom-left
    ]
    polygon = patches.Polygon(points, closed=True, color="#ea3e21")
    ax.add_patch(polygon)

    # Inner white polygon
    inner_points = [
        (4.7, 4),   # Bottom-left
        (4.7, 10),  # Top-left
        (5.3, 9.8), # Top-right
        (5.3, 3),   # Bottom-right
    ]
    inner_polygon = patches.Polygon(inner_points, closed=True, color="white")
    ax.add_patch(inner_polygon)

    plt.text(2, 14, "Logo Microsoft Office", fontsize=14, ha='left')

    plt.show()

if __name__ == "__main__":
    draw_microsoft_office_logo()
```
